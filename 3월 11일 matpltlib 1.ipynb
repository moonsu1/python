# matplotlib 패키지 (시각화)

```python
import matplotlib as mpl
import matplotlib.pylab as plt
```

```python
%matplotlib inline
#이 명령어를 써줘야 주피터 노트북에서 그림이 그려진다
```

```python
#line plot
plt.title("Plot")
plt.plot([1,2,3,5])
plt.show
#데이터가 시간 순서에 따라 어떻게 변화하는지 보여줌
```

```python
#x축의 범위를 별도로 지정하고 싶을 때
#line plot
plt.title("x축의 tick 위치를 명시")
plt.plot([10,20,30,40],[1,2,3,6])
plt.show()
```

# 스타일 지정

```python
plt.title("'rs--' 스타일의 plot ")
plt.plot([10,20,30],[1,2,3], 'rs--')
plt.show()
#순서대로 색깔, 마커, 선종류를 뜻함
#- -는 점선으로 바꿔준다. r은 red s는 사각형 marker을 의미
#여러가지 종류는 외우기 보다 찾아가면서 하면 될 듯하다.
```

```python
#c 색, lw 선굵기, ls 선스타일, marker 마커종류, ms 마커크기, mec 마커선색깔, mew 마커 선 굵기, mfc 마커 내부색깔
```

```python
#앞의 세가지 종류는 기본이고 만약에 다른 것을 쓰고 싶다면 약어 말고 full로 적어야한다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.show()
```

# 그림 범위 지정

```python
#그림이 빽빽하게 들어차 있으면 그림이 범위 경계면에 가려서 잘 보이지 않을수 있다
#그럴 때는 xlim, ylim을 
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

```python
import numpy as np
#아까 앞에서 x tick이라는 것을 설정했는데 
#x tick과 y tick을 따로 지정 해주려면 이렇게 쓴다.
#예시는 분포로 지정해주지 않으면 끝까지 이어질 수 있다.
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
#위 분포에서 tick을 추가 할 때 수학식을 넣을 수도 있다.
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

```python
#위에서는 보이지 않지만 원래 그래프에 grid line이 그려져 있다 . 보기 싫으면 삭제 할 수 있다
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("Grid 제거")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid(False) #해당부분
plt.show()
```

# 한 그래프에 여러개의 선 그리기

```python
#만약에 한 그래프에 여러개의 선을 그리고 싶으면 (x,y,스타일)의 형태를 반복하면 여러개가 그려진다
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
#여기서는 t,t     t,0.5*t**2     t,0.2*t**3 이렇게 3개의 선이 들어가 있다
plt.show()
```

# 여러개의 그래프를 한그래프로 합쳐 쓰기

```python
#두개 이상의 그래프를 합쳐서 한번에 표시하고 싶다면 #plt.hold(True)를 쓴다
#하지만 요즘에는 plt.hold(true)가 포함 되어 있어서 두개 연속으로 쓰면 알아서 적용된다. 한 칸안에 쓰는데 따로 쓰고 싶다면 #plt.hold(False)삽입
plt.title("복수의 plot 명령을 한 그림에서 표현")
plt.plot([1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.hold(False)   # <- 따로 그리기
plt.plot([9, 16, 4, 1],
         c="k", lw=3, ls=":", marker="s", ms=10, mec="m", mew=5, mfc="c")

plt.show()
```

# 범례 지정

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")


plt.plot(X, C, ls="--", label="cosine")
plt.hold(True)
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
#각 선이 어떤 선인지 주석 느낌으로 붙이고 싶다면 label 옵션을 사용하면 된다
```

# x축과 y축의 label 표시하기

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")

plt.xlabel("time")      #이 부분을 써주면 축마다 이름이 생긴다
plt.ylabel("amplitude")

plt.title("Cosine Plot")
plt.show()
```
